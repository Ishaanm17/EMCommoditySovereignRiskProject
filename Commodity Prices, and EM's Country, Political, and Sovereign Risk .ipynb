```python
EMCPRML=pd.get_dummies(EMCPRML, columns=[' Impact '])
```

```python
EMCPRML
```

```python
EMCPRML.columns
```

```python
EMCPRML=EMCPRML.rename(columns={' Impact _High':'Impact_High','Category of Political Risk according to ICRG ':'PoliticalRisk_CategoryICRG','Category of SovereignRisk according to ICRG ':'SovereignRisk_CategoryICRG','Category of Country  Risk according to ICRG ':'CountryRisk_CategoryICRG'})
```

```python
EMCPRML.Impact_High.value_counts()
```

# Political Risk and Commodity Prices Machine Learning 

```python
from patsy import dmatrices
y,X = dmatrices('Impact_High ~ PoliticalRisk_CategoryICRG',data=EMCPRML)
```

```python
 y=np.ravel(y)
```

```python
from sklearn.linear_model import LogisticRegression as logit 
logit(solver='lbfgs').fit(X,y).score(X,y)
```

```python
 from sklearn.model_selection import cross_val_score
cross_val_score(logit(solver='lbfgs'),X,y,cv=5,scoring='accuracy').mean()
```

```python
 cross_val_score(logit(solver='lbfgs'),X,y,cv=3,scoring='roc_auc').mean()
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn 
cross_val_score(knn(),X,y,cv=3,scoring='accuracy').mean()
```

```python
from sklearn.ensemble import RandomForestClassifier as rf 
cross_val_score(rf(n_estimators=100, max_depth=2,max_features=2),X,y,cv=2,scoring='accuracy').mean()
```

```python
for i in range(1,10):    
        for j in range(1,3):        
            print(i,j,cross_val_score(rf(n_estimators=100,max_depth=i,max_features =j),X,y,cv=3,scoring='accuracy').mean())
```

```python
EMCPRML['high_impact_rf'] = rf(n_estimators=100, max_depth=2,max_features=2).fit( X,y).predict(X)
EMCPRML.high_impact_rf.value_counts()
```

#### This model shows that from 2017-2023, Country Risk will have a 100% chance of highly impacting commodity prices.
